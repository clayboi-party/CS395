```python
import requests
import json
from urllib.parse import urljoin
from bs4 import BeautifulSoup

# base URL
base_url = "https://www.haberturk.com"

# load URLs
with open('urls.json', 'r') as file:
    data = json.load(file)
    urls = data['unique_urls']

# list to hold the results
results = []

# go through the URLs to get their meta tags
total_urls = len(urls)
for i, url in enumerate(urls, start=1):
    full_url = urljoin(base_url, url)
    try:
        # fetch URL content
        response = requests.get(full_url)
        
        # parse HTML content
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # extract meta tags
        meta_tags = soup.find_all('meta')
        meta_info = [{'name': tag.get('name'), 'content': tag.get('content')} for tag in meta_tags if tag.get('name')]

        # progress print
        print(f"{i}/{total_urls} - {full_url}: Meta tags fetched")
        
        # append result to the list
        results.append({'url': full_url, 'meta_tags': meta_info})
    except requests.RequestException as e:
        # for request errors
        print(f"{i}/{total_urls} - {full_url}: Failed to retrieve meta tags ({e})")
        results.append({'url': full_url, 'meta_tags': None, 'error': str(e)})

# save the results to JSON file
with open('url_meta_tags.json', 'w', encoding='utf-8') as json_file:
    json.dump(results, json_file, indent=4)

# save the results to HTML file
html_content = "<html><head><title>URL Meta Tags</title></head><body><table border='1'>"
html_content += "<tr><th>URL</th><th>Meta Tags</th></tr>"

for result in results:
    if result['meta_tags'] is not None:
        meta_tags_str = "<br>".join([f"{tag['name']}: {tag['content']}" for tag in result['meta_tags']])
    else:
        meta_tags_str = "N/A"
    html_content += f"<tr><td>{result['url']}</td><td>{meta_tags_str}</td></tr>"

html_content += "</table></body></html>"

with open('url_meta_tags.html', 'w', encoding='utf-8') as html_file:
    html_file.write(html_content)

print("Data saved to url_meta_tags.json and url_meta_tags.html")
```
