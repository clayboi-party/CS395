```python
import json
import pandas as pd

json_file_path = 'urls.json'

with open(json_file_path,'r') as file:
    data = json.load(file)

internal_urls = data['internal_urls']
relative_urls = data['relative_urls']
external_urls = data['external_urls']
absolute_urls = data['absolute_urls']
unique_external_urls = data['unique_external_urls']
total_url_list = data['total_url_list']
unique_urls = data['unique_urls']
relative_urls = data['relative_urls']

# create dataFrames
df_internal = pd.DataFrame(internal_urls, columns=['Internal URLs'])
df_relative = pd.DataFrame(relative_urls, columns=['Relative URLs'])
df_external = pd.DataFrame(external_urls, columns=['External URLs'])
df_absolute = pd.DataFrame(absolute_urls, columns=['Absolute URLs'])
df_unique_external = pd.DataFrame(unique_external_urls, columns=['Unique External URLs'])
df_total_url = pd.DataFrame(total_url_list, columns=['Total URL List'])
df_unique = pd.DataFrame(unique_urls, columns=['Unique URLs'])

# export dataFrames to html and save to files
def export_to_html(df, filename):
    html_data = df.to_html(index=False)
    with open(filename, 'w') as file:
        file.write(html_data)
    print(f"Exported to {filename}")

export_to_html(df_internal, 'df_internal.html')
export_to_html(df_absolute, 'df_absolute.html')
export_to_html(df_relative, 'df_relative.html')
export_to_html(df_unique, 'df_unique.html')
export_to_html(df_external, 'df_external.html')
export_to_html(df_unique_external, 'df_unique_external.html')
export_to_html(df_total_url, 'df_total_url.html')

print("\nTotal URL Count:", len(total_url_list))

print(f"\nTotal unique URLs: {len(df_unique)}")

print("\nTotal Internal URLs:", len(internal_urls))

print("\nTotal Absolute URLs:", len(absolute_urls))


print(f"\nTotal relative URLs: {len(relative_urls)}")

print(f"\nTotal external URLs: {len(df_external)}")

print (f"\nTotal unique external URLs: {len(df_unique_external)}")

print("\nInternal URLs:")
for url in internal_urls:
    print(url)

print ()
print ()

print("Absolute URLs:")
for url in absolute_urls:
    print (url)

print ()
print ()

print("Relative URLs:")
for url in df_relative['Relative URLs']:
    print (url)


print ()
print ()

print("\nUnique URLs:")
for url in df_unique['Unique URLs']:
    print(url)

print ()
print ()

print("External URLs:")
for url in df_external['External URLs']:
    print(url)

print ()
print ()

print("Unique External URLs")
for url in df_unique_external['Unique External URLs']:
    print(url)
```
